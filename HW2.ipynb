```python
import numpy as np; from scipy import stats; import matplotlib.pyplot as plt

# Part II
# 1.

# Set the seed
np.random.seed(1005810605)

n = 3; x = stats.norm(loc=0,scale=1).rvs(n); theta_0,tau = 4,1/2; parameter_domain = np.linspace(-10,10,1001);

likelihood = []
for xi in x:
    likelihood.append(stats.norm.pdf(xi, loc=parameter_domain, scale=1))

prior = stats.norm(loc=theta_0, scale=1/tau**0.5).pdf(parameter_domain) 
joint = np.prod(likelihood)*prior 
posterior = stats.norm(loc=(tau*theta_0+x.sum())/(tau+n), scale=1/(tau+n)**0.5).pdf(parameter_domain)

plt.figure(figsize=(16,5))
plt.plot(parameter_domain, likelihood[0], label="likelihood 1")
plt.plot(parameter_domain, likelihood[1], label="likelihood 2")
plt.plot(parameter_domain, likelihood[2], label="likelihood 3")
plt.plot(parameter_domain, prior, label="prior")
plt.plot(parameter_domain, posterior, label="posterior")
plt.plot(parameter_domain, joint, linestyle="--", label="joint")
plt.title("Visualize prior, likelihoods, posteriors and joint distribution")
plt.legend()
```

```python
# 2.
# Set the seed
np.random.seed(1005810605)
n = 3; x = stats.norm(loc=0,scale=1).rvs(n); theta_0,tau = 4,1/4; parameter_domain = np.linspace(-10,10,1001); 
likelihood = []
for xi in x:
    likelihood.append(stats.norm.logpdf(xi, loc=parameter_domain, scale=1))
prior = stats.norm(loc=theta_0, scale=1/tau**0.5).logpdf(parameter_domain)
joint = sum(likelihood)+prior
posterior = stats.norm(loc=(tau*theta_0+x.sum())/(tau+n), scale=1/(tau+n)**0.5).logpdf(parameter_domain)
plt.figure(figsize=(16,5))
plt.plot(parameter_domain, likelihood[0], label="loglikelihood1")
plt.plot(parameter_domain, likelihood[1], label="loglikelihood2")
plt.plot(parameter_domain, likelihood[2], label="loglikelihood3")
plt.plot(parameter_domain, prior, label="logprior")
plt.plot(parameter_domain, posterior, label="logposterior")
plt.plot(parameter_domain, joint, linestyle="--", label="logjoint")
plt.title("Visualize logprior, loglikelihoods, logposteriors and logjoint distribution")
plt.legend()
```

The posterior distribution of a normal likelihood with $p(\theta) \propto 1$ is still the normal distribution.

4.
The difference between likelihoods and posteriors is that the likelihood is the probability of the observed data given parameters while the posterior is the proportional probability of the parameters given both the observed data and the prior. The marginal likelihood constant is used to connect the likelihood and the posterior, and its exact value is not needed to calculated when the priors are improper.
